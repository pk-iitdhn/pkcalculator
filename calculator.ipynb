#   CALCULATOR

```python
from tkinter import *
from tkinter import messagebox

win = Tk()
win.geometry("345x540")
win.resizable(False,False)
win.configure(bg="white")
win.title("pkCalculator")
global brct
brct = False

def ac():
    exp.delete(0,END)

def equal():
    e = exp.get()
    result = eval(e)
    exp.delete(0,END)
    exp.insert(0,result)

def backspace():
    xp = exp.get()
    exp.delete(0,"end")
    exp.insert(0,xp[:-1])

def expression(num):
    s = exp.get()
    s=s+num
    exp.delete(0,"end")
    exp.insert(0,s)



def bracket():
        global brct
        if brct:
            brct=False
            expression(")")
        else:            
            brct=True
            expression("(")

entryframe = Frame(win)
entryframe.configure(bg="snow",bd=5)
entryframe.place(relx=0,rely=0,relheight=0.26,relwidth=1)

exp = Entry(entryframe,font=("monospace",28,"bold"),bg="gray98",fg="blue",bd=2,justify=RIGHT)
exp.place(relx=0,rely=0,relheight=1,relwidth=1)

buttonframe = Frame(win)
buttonframe.configure(bg="snow",bd=2)
buttonframe.place(relheight=0.75,relwidth=1,relx=0,rely=0.251)

bac = Button(buttonframe,text="AC",font=("Arial",18,"bold"),bd=2,fg="black",bg="grey",width=5,height=2,command=ac)
bac.grid(row=0,column=0)
bbrckt = Button(buttonframe,text="( )",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=bracket)
bbrckt.grid(row=0,column=1)
bmod = Button(buttonframe,text="%",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=lambda:expression("%"))
bmod.grid(row=0,column=2)
bdiv = Button(buttonframe,text="/",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=lambda:expression("/"))
bdiv.grid(row=0,column=3)

b1 = Button(buttonframe,text="1",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("1"))
b1.grid(row=1,column=0)
b2 = Button(buttonframe,text="2",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("2"))
b2.grid(row=1,column=1)
b3 = Button(buttonframe,text="3",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("3"))
b3.grid(row=1,column=2)
badd = Button(buttonframe,text="+",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=lambda:expression("+"))
badd.grid(row=1,column=3)
b4 = Button(buttonframe,text="4",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("4"))
b4.grid(row=2,column=0)
b5 = Button(buttonframe,text="5",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("5"))
b5.grid(row=2,column=1)
b6 = Button(buttonframe,text="6",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("6"))
b6.grid(row=2,column=2)
bsub = Button(buttonframe,text="-",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=lambda:expression("-"))
bsub.grid(row=2,column=3)
b7 = Button(buttonframe,text="7",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("7"))
b7.grid(row=3,column=0)
b8 = Button(buttonframe,text="8",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("8"))
b8.grid(row=3,column=1)
b9 = Button(buttonframe,text="9",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("9"))
b9.grid(row=3,column=2)
bmul = Button(buttonframe,text="x",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=lambda:expression("*"))
bmul.grid(row=3,column=3)
bdot = Button(buttonframe,text=".",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("."))
bdot.grid(row=4,column=0)
b0 = Button(buttonframe,text="0",font=("Arial",18,"bold"),bd=2,fg="black",bg="light grey",width=5,height=2,command=lambda:expression("0"))
b0.grid(row=4,column=1)
bbksp = Button(buttonframe,text="CLR",font=("Arial",18,"bold"),bd=2,fg="black",bg="grey",width=5,height=2,command=backspace)
bbksp.grid(row=4,column=2)
beq = Button(buttonframe,text="=",font=("Arial",18,"bold"),bd=2,fg="white",bg="grey",width=5,height=2,command=equal)
beq.grid(row=4,column=3)

mainloop()
```
